```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.linspace(0,1,5)
y=x**2
```

```python
plt.plot(x,y,'r-')
plt.xlabel('Xlabel')
plt.ylabel('Ylabel')
plt.title('this is sample graph')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

#Another method to creata plots

```python
fig=plt.figure()
axes=fig.add_axes([0.4,0.4,0.8,0.8])
axes.set_xlabel('Xlabel')
axes.set_ylabel('Ylabe')
axes.set_title('sample graph')
axes.plot(x,y)

```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.set_xlabel('Xlabel')
axes.set_ylabel('Ylabe')
axes.set_title('sample graph')
axes.plot(x,y)

axes1=fig.add_axes([0.2,0.5,0.4,0.3])
axes1.set_xlabel('Xlabel_inner')
axes1.set_ylabel('Ylabe_innter')
axes1.set_title('sample graph_inner')
axes1.plot(y,x)
```
