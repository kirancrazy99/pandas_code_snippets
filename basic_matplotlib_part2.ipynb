# Plot Appearance

```python
fig=plt.figure()

axes=fig.add_axes([0,0,1,1])

x=np.linspace(0,5,11)
y=x**2

axes.plot(x,y,color='red')
```

```python
fig=plt.figure()

axes=fig.add_axes([0,0,1,1])

x=np.linspace(0,5,11)
y=x**2

axes.plot(x,y,color='red',linewidth=2)
```

```python
fig=plt.figure()

axes=fig.add_axes([0,0,1,1])

x=np.linspace(0,5,11)
y=x**2

axes.plot(x,y,color='red',linewidth=2,linestyle='--')
```

```python
fig=plt.figure()

axes=fig.add_axes([0,0,1,1])

x=np.linspace(0,5,11)
y=x**2

axes.plot(x,y,color='red',linewidth=2,linestyle='--',marker='o',markersize=10)
```

```python
fig=plt.figure()

axes=fig.add_axes([0,0,1,1])

x=np.linspace(0,5,11)
y=x**2

axes.plot(x,y,color='red',linewidth=2,linestyle='--',marker='o',markersize=10,markerfacecolor='yellow')
```

```python
fig=plt.figure()

axes=fig.add_axes([0,0,1,1])

x=np.linspace(0,5,11)
y=x**2

axes.plot(x,y,color='red',linewidth=2,linestyle='--',marker='o',markersize=10,markerfacecolor='yellow',
         markeredgewidth=3,markeredgecolor='green')
```
