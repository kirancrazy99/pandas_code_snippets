```python
x=np.linspace(0,5,11)
y=x**2
```

```python
fig=plt.figure()

axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes1=fig.add_axes([0.2,0.5,0.4,0.3])


axes.set_xlabel('xlable_outerplot')
axes.set_ylabel('ylable_outperplot')
axes.set_title('Outerplot')

axes1.set_xlabel('xlabel_innerplot')
axes1.set_ylabel('ylabel_innerplot')
axes1.set_title('innerpot')


axes.plot(x,y)
axes1.plot(y,x)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('first plot')

axes[1].plot(y,x)
axes[1].set_title('sample plot')
```

# Fig Size and DPI(Dots per Inch)

```python
fig=plt.figure(figsize=(8,2))
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig,axes=plt.subplots(figsize=(8,4),nrows=2,ncols=1)
axes[0].plot(x,y)
axes[0].set_title('first_plot')

axes[1].plot(y,x)
axes[1].set_title('second plot')
plt.tight_layout()
```
